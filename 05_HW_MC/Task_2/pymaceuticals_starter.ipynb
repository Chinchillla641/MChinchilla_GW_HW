```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_drug_data_to_load = "D:/Git/GWU-ARL-DATA-PT-09-2019-U-C/02-Homework/05-Matplotlib/Instructions/Part-2/Pymaceuticals/data/mouse_drug_data.csv"
clinical_trial_data_to_load = "D:/Git/GWU-ARL-DATA-PT-09-2019-U-C/02-Homework/05-Matplotlib/Instructions/Part-2/Pymaceuticals/data/clinicaltrial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data
mouse_dta = pd.read_csv(mouse_drug_data_to_load)

clinic_dta = pd.read_csv(clinical_trial_data_to_load)

# Combine the data into a single dataset
df =  pd.merge(clinic_dta, mouse_dta, on = "Mouse ID", how = "left")

# Display the data table for preview
df
```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
mean_tumor = df.groupby(["Drug", "Timepoint"]).mean()["Tumor Volume (mm3)"]

# Convert to DataFrame
mtumor_df = pd.DataFrame(mean_tumor)

# Preview DataFrame
mtumor_df
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
mean_tumor = df.groupby(["Drug", "Timepoint"]).sem()["Tumor Volume (mm3)"]

# Convert to DataFrame
setumor_df = pd.DataFrame(mean_tumor)

# Preview DataFrame
setumor_df
```

```python
# Minor Data Munging to Re-Format the Data Frames
mtumor_df.reset_index(inplace = True)
setumor_df.reset_index(inplace = True)

mtumor_df = mtumor_df.pivot(index = "Timepoint", columns = "Drug")["Tumor Volume (mm3)"]
setumor_df = setumor_df.pivot(index = "Timepoint", columns = "Drug")["Tumor Volume (mm3)"]

# Preview that Reformatting worked
mtumor_df
setumor_df
```

```python
# Generate the Plot (with Error Bars)
plt.errorbar(mtumor_df.index, mtumor_df["Capomulin"], yerr = setumor_df["Capomulin"])
plt.errorbar(mtumor_df.index, mtumor_df["Ceftamin"], yerr = setumor_df["Ceftamin"])
plt.errorbar(mtumor_df.index, mtumor_df["Infubinol"], yerr = setumor_df["Infubinol"])
plt.errorbar(mtumor_df.index, mtumor_df["Ketapril"], yerr = setumor_df["Ketapril"])
plt.errorbar(mtumor_df.index, mtumor_df["Naftisol"], yerr = setumor_df["Naftisol"])
plt.errorbar(mtumor_df.index, mtumor_df["Placebo"], yerr = setumor_df["Placebo"])
plt.errorbar(mtumor_df.index, mtumor_df["Propriva"], yerr = setumor_df["Propriva"])
plt.errorbar(mtumor_df.index, mtumor_df["Ramicane"], yerr = setumor_df["Ramicane"])
plt.errorbar(mtumor_df.index, mtumor_df["Stelasyn"], yerr = setumor_df["Stelasyn"])
plt.errorbar(mtumor_df.index, mtumor_df["Zoniferol"], yerr = setumor_df["Zoniferol"])

plt.title("Drug Efficacy")
# Save the Figure

plt.savefig("plot_error.jpeg")
```

```python
# Show the Figure
plt.show()
```

## Metastatic Response to Treatment

```python
# Store the Mean Met. Site Data Grouped by Drug and Timepoint 
mmet = df.groupby(["Drug", "Timepoint"]).mean()["Metastatic Sites"]

# Convert to DataFrame
mmet = pd.DataFrame(mmet)

# Preview DataFrame
mmet
```

```python
# Store the Standard Error associated with Met. Sites Grouped by Drug and Timepoint 
semet = df.groupby(["Drug", "Timepoint"]).sem()["Metastatic Sites"]

# Convert to DataFrame
semet = pd.DataFrame(semet)

# Preview DataFrame
semet
```

```python
# Minor Data Munging to Re-Format the Data Frames
mmet.reset_index(inplace = True)
semet.reset_index(inplace = True)

mmet = mmet.pivot(index = "Timepoint", columns = "Drug")["Metastatic Sites"]
semet = semet.pivot(index = "Timepoint", columns = "Drug")["Metastatic Sites"]

# Preview that Reformatting worked
mmet
semet
```

```python
# Generate the Plot (with Error Bars)
plt.errorbar(mmet.index, mmet["Capomulin"], yerr = semet["Capomulin"])
plt.errorbar(mmet.index, mmet["Ceftamin"], yerr = semet["Ceftamin"])
plt.errorbar(mmet.index, mmet["Infubinol"], yerr = semet["Infubinol"])
plt.errorbar(mmet.index, mmet["Ketapril"], yerr = semet["Ketapril"])
plt.errorbar(mmet.index, mmet["Naftisol"], yerr = semet["Naftisol"])
plt.errorbar(mmet.index, mmet["Placebo"], yerr = semet["Placebo"])
plt.errorbar(mmet.index, mmet["Propriva"], yerr = semet["Propriva"])
plt.errorbar(mmet.index, mmet["Ramicane"], yerr = semet["Ramicane"])
plt.errorbar(mmet.index, mmet["Stelasyn"], yerr = semet["Stelasyn"])
plt.errorbar(mmet.index, mmet["Zoniferol"], yerr = semet["Zoniferol"])

plt.title("Metastatic Sites")
# Save the Figure
plt.savefig("met_error.jpeg")

# Show the Figure
plt.show()
```

## Survival Rates

```python
# Store the Count of Mice Grouped by Drug and Timepoint (W can pass any metric)
scount = df.groupby(["Drug", "Timepoint"]).count()["Tumor Volume (mm3)"]

# Convert to DataFrame
scount = pd.DataFrame(scount)

# Preview DataFrame
scount
```

```python
# Minor Data Munging to Re-Format the Data Frames
scount.reset_index(inplace = True)
scount.columns = ["Drug", "Timepoint", "Survival Count"]

scount = scount.pivot(index = "Timepoint", columns = "Drug")["Survival Count"]

# Preview the Data Frame
scount

```

```python
# Generate the Plot (Accounting for percentages)
plt.plot(100 * scount["Capomulin"] / 25)
plt.plot(100 * scount["Ceftamin"] / 25)
plt.plot(100 * scount["Infubinol"] / 25)
plt.plot(100 * scount["Ketapril"] / 25)
plt.plot(100 * scount["Naftisol"] / 25)
plt.plot(100 * scount["Placebo"] / 25)
plt.plot(100 * scount["Propriva"] / 25)
plt.plot(100 * scount["Ramicane"] / 25)
plt.plot(100 * scount["Stelasyn"] / 25)
plt.plot(100 * scount["Zoniferol"] / 25)

plt.title("% Survival by Drug")
# Save the Figure
plt.savefig("pct.jpeg")

# Show the Figure
plt.show()
```
