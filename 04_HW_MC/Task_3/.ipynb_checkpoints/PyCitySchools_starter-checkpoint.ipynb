```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "D:/Git/GWU-ARL-DATA-PT-09-2019-U-C/02-Homework/04-Pandas/Instructions/Part-3/Task-B-PyCitySchools/PyCitySchools/Resources/schools_complete.csv"
student_data_to_load = "D:/Git/GWU-ARL-DATA-PT-09-2019-U-C/02-Homework/04-Pandas/Instructions/Part-3/Task-B-PyCitySchools/PyCitySchools/Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
school_ct = school_data_complete.loc[:, ["school_name"]]
school_ct = pd.DataFrame.drop_duplicates(school_ct)
school_count = school_ct["school_name"].count()
school_ct = pd.DataFrame({"Total Schools": [school_count]})

std_ct = school_data_complete.loc[:, ["Student ID"]]
std_ct = pd.DataFrame.drop_duplicates(std_ct)
std_count = std_ct["Student ID"].count()
std_ct = pd.DataFrame({"Total Students": [std_count]})

totbudget = school_data_complete.loc[:, ["school_name", "budget"]]
totbudget = pd.DataFrame.drop_duplicates(totbudget)
totbudgett = totbudget["budget"].sum()
totbudget = pd.DataFrame({"Total Budget": [totbudgett]})

avg_math = school_data_complete["math_score"].mean()
avg_math_score = pd.DataFrame({"Average Math Score": [avg_math]})

avg_read = school_data_complete["reading_score"].mean()
avg_read_score = pd.DataFrame({"Average Reading Score": [avg_read]})

school_data_pass = school_data_complete
school_data_pass["mathpass"] = school_data_pass["math_score"] >= 70
school_data_pass["readpass"] = school_data_pass["reading_score"] >= 70
math_pass = 100 * school_data_pass["mathpass"].mean()
read_pass = 100 * school_data_pass["readpass"].mean()
over_pass = (math_pass + read_pass) / 2

pass_data = pd.DataFrame({"% Passing Math": [math_pass],
                          "% Passing Reading": [read_pass],
                          "% Overall Passing Rate": [over_pass]
                         })

dist_summ = pd.merge(school_ct, std_ct, left_index=True, right_index=True)
dist_summ = pd.merge(dist_summ, totbudget, left_index=True, right_index=True)
dist_summ = pd.merge(dist_summ, avg_math_score, left_index=True, right_index=True)
dist_summ = pd.merge(dist_summ, avg_read_score, left_index=True, right_index=True)
dist_summ = pd.merge(dist_summ, pass_data, left_index=True, right_index=True)
dist_summ
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)
  
* Create a dataframe to hold the above results

```python
dist = school_data.loc[:, ["school_name", "type", "size", "budget"]]
dist["std_budget"] = dist["budget"] / dist["size"]

avg_mathschool = school_data_complete["math_score"].groupby(school_data_complete["school_name"]).mean()
avg_maths = pd.DataFrame(avg_mathschool)
avg_readschool = school_data_complete["reading_score"].groupby(school_data_complete["school_name"]).mean()
avg_reads = pd.DataFrame(avg_readschool)

math_passs = 100 * school_data_pass["mathpass"].groupby(school_data_pass["school_name"]).mean()
math_pass_school = math_passs
read_passs = 100 * school_data_pass["readpass"].groupby(school_data_pass["school_name"]).mean()
read_pass_school = read_passs

pass_datas = pd.merge(math_pass_school, read_pass_school, on = "school_name", how = "inner")
pass_datas["overall"] = (pass_datas["mathpass"] + pass_datas["readpass"]) / 2

dist = pd.merge(dist, avg_mathschool, on = "school_name", how = "inner")
dist = pd.merge(dist, avg_readschool, on = "school_name", how = "inner")
dist = pd.merge(dist, pass_datas, on = "school_name", how = "inner")
dist.columns = ["School Name", "School Type", "Total Students", \
                "Total School Budget", "Per Student Budget", \
                "Average Math Score", "Average Reading Score", 
                "% Passing Math", "% Passing Reading", "% Overall Passing Rate"]
```

## Math Scores by Grade

* Create a table that lists the average Reading Score for students of each grade level (9th, 10th, 11th, 12th) at each school.

  * Create a pandas series for each grade. Hint: use a conditional statement.
  
  * Group each series by school
  
  * Combine the series into a dataframe
  
  * Optional: give the displayed data cleaner formatting

```python
dta_9 = school_data_complete.loc[school_data_complete["grade"] == "9th", :]
dta_9 = dta_9["math_score"].groupby(school_data_complete["school_name"]).mean()
dta_9 = pd.DataFrame(dta_9)
dta_9.reset_index(inplace = True)
dta_9.columns = ["school_name", "9th"] 

dta_10 = school_data_complete.loc[school_data_complete["grade"] == "10th", :]
dta_10 = dta_10["math_score"].groupby(school_data_complete["school_name"]).mean()
dta_10 = pd.DataFrame(dta_10)
dta_10.reset_index(inplace = True)
dta_10.columns = ["school_name", "10th"]

dta_11 = school_data_complete.loc[school_data_complete["grade"] == "11th", :]
dta_11 = dta_11["math_score"].groupby(school_data_complete["school_name"]).mean()
dta_11 = pd.DataFrame(dta_11)
dta_11.reset_index(inplace = True)
dta_11.columns = ["school_name", "11th"]

dta_12 = school_data_complete.loc[school_data_complete["grade"] == "12th", :]
dta_12 = dta_12["math_score"].groupby(school_data_complete["school_name"]).mean()
dta_12 = pd.DataFrame(dta_12)
dta_12.reset_index(inplace = True)
dta_12.columns = ["school_name", "12th"]

math_by_grade = pd.merge(dta_9, dta_10, on = "school_name", how = "inner")
math_by_grade = pd.merge(math_by_grade, dta_11, on = "school_name", how = "inner")  
math_by_grade = pd.merge(math_by_grade, dta_12, on = "school_name", how = "inner")
math_by_grade
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
# Sample bins. Feel free to create your own bins.
spending_bins = [0, 585, 615, 645, 675]
group_names = ["<$585", "$585-615", "$615-645", "$645-675"]

dist["Spending Ranges (Per Student)"] = pd.cut(dist["Per Student Budget"],spending_bins, labels = group_names)
avgm_bin = dist["Average Math Score"].groupby(dist["Spending Ranges (Per Student)"]).mean()
avgm_bin = pd.DataFrame(avgm_bin)

avgr_bin = dist["Average Reading Score"].groupby(dist["Spending Ranges (Per Student)"]).mean()
avgr_bin = pd.DataFrame(avgr_bin)

pctm_bin = dist["% Passing Math"].groupby(dist["Spending Ranges (Per Student)"]).mean()
pctm_bin = pd.DataFrame(pctm_bin)

pctr_bin = dist["% Passing Reading"].groupby(dist["Spending Ranges (Per Student)"]).mean()
pctr_bin = pd.DataFrame(pctr_bin)

pcto_bin = dist["% Overall Passing Rate"].groupby(dist["Spending Ranges (Per Student)"]).mean()
pcto_bin = pd.DataFrame(pcto_bin)

school_spend_bin = pd.merge(avgm_bin, avgr_bin, on = "Spending Ranges (Per Student)", how = "inner")
school_spend_bin = pd.merge(school_spend_bin, pctm_bin, on = "Spending Ranges (Per Student)", how = "inner")
school_spend_bin = pd.merge(school_spend_bin, pctr_bin, on = "Spending Ranges (Per Student)", how = "inner")
school_spend_bin = pd.merge(school_spend_bin, pcto_bin, on = "Spending Ranges (Per Student)", how = "inner")
school_spend_bin
```
