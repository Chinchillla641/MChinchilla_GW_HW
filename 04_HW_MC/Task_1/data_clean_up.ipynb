```python
# Dependencies
import pandas as pd
import numpy as np
```

```python
# load CSV
df = pd.read_csv("Resources/2016-FCC-New-Coders-Survey-Data.csv", low_memory = False)
df.head()
```

```python
# Read with pandas
```

```python
# Inspect all columns
df.count()
```

```python
df.shape
```

```python
# Extract only columns 0, 1, 2, 3, 4, 7, 8, 9, 10, 11, 29, 30, 32, 36, 37, 45, 48, 56, 110, 111
columns = [0, 1, 2, 3, 4, 7, 8, 9, 10, 11, 29, 30, 32, 36, 37, 45, 48, 56, 110, 111]
extract_df = df.iloc[:, columns]
extract_df.head()
```

```python
# Change "0" to "No" and "1" to "Yes" in response columns
extract_df['AttendedBootcamp'] = extract_df['AttendedBootcamp'].apply(lambda x: "no" if x==0 else "yes")
extract_df.head()
```

```python
# Calculate total number of respondents in survey
total_respondents = extract_df.shape[0]
```

```python
# Extract rows corresponding only to people who attended a bootcamp
attend_df = extract_df.loc[extract_df["AttendedBootcamp"] == "yes", :]
attend_df.head()
```

```python
# Calculate average age of attendees
mean_age = attend_df["Age"].mean()
# Calculate how many people attended a bootcamp
total_attend = attend_df.shape[0]

# Calculate how many attendees hold degrees
attend_df["Degree"] = pd.Series("Yes", index=attend_df.index)

attend_df.loc[attend_df.SchoolDegree == "some college credit, no degree", "Degree"] = "No"
attend_df.loc[attend_df.SchoolDegree == "high school diploma or equivalent (GED)", "Degree"] = "No"
attend_df.loc[attend_df.SchoolDegree == "trade, technical, or vocational training", "Degree"] = "No"
attend_df.loc[attend_df.SchoolDegree == "some high school", "Degree"] = "No"
attend_df.loc[attend_df.SchoolDegree == "no high school (secondary school)", "Degree"] = "No"

coll_degree = attend_df["Degree"].value_counts(dropna = False)["Yes"]

# Count number of attendees who self-identify as male; female; or are of non-binary gender identification
male = attend_df["Gender"].value_counts()["male"]
female = attend_df["Gender"].value_counts()["female"]
nangen = len(attend_df) - attend_df['Gender'].count()
other = total_attend - (male + female + nangen)

# Calculate percentage of respondents who attended a bootcamp
pct_attend = total_attend / total_respondents

# Calculate percentage of respondents belonging to each gender
pct_male = 100 * male / total_attend
pct_female = 100 * female / total_attend
pct_other = 100 * other / total_attend

# Calculate percentage of attendees with a school degree
pct_degree = 100 * coll_degree / total_attend

# Calculate average post-bootcamp salary
mean_sal = attend_df["BootcampPostSalary"].mean()
```

```python
# Create a new table consolodating above calculations
table = pd.DataFrame = ({"Total Surveyed": [total_respondents],
                        "Total Bootcamp Atendees": [total_attend],
                        "% attneded bootcamp": [pct_attend],
                        "Avg. Age": [mean_age],
                        "% Male": [male],
                        "% Female": [female],
                        "% Non Gender": [other],
                        "Has a degree": [pct_degree],
                        "Average Post Bootcamp Salary": mean_sal})

# table = table[["Total Surveyed", "Total Bootcamp Atendees", "% attneded bootcamp", "Avg. Age", "% Male", "% Female", \
#                "% Non Gender", "Has a degree", "Average Post Bootcamp Salary"]]
table.round(2)
```
