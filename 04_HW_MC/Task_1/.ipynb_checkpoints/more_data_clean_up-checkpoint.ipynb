```python
# Dependencies
import os
import pandas as pd
import numpy as np
```

```python
# load CSV
df = pd.read_csv("D:/Git/GWU-ARL-DATA-PT-09-2019-U-C/02-Homework/04-Pandas/Instructions/Part-1/Pandas_Mini_Project_Task_A/Resources/2016-FCC-New-Coders-Survey-Data.csv", low_memory = False)
df.shape
```

```python
# Read with pandas--low_memory required to suppress errors about mixed data types
```

```python
# Take only columns 0, 1, 2, 3, 4, 6, 7, 8, 9, 10, 11, 29, 30, 32, 36, 37, 45, 48, 56, 110, 111
columns = [0, 1, 2, 3, 4, 6, 7, 8, 9, 10, 11, 29, 30, 32, 36, 37, 45, 48, 56, 110, 111]
extract_df = df.iloc[:, columns]
extract_df.shape
```

```python
df.count()
```

```python
# Change reading 0 and 1 to No and Yes, respectively
extract_df['AttendedBootcamp'] = extract_df['AttendedBootcamp'].apply(lambda x: "no" if x==0 else "yes")
extract_df['BootcampRecommend'] = extract_df['BootcampRecommend'].apply(lambda x: "no" if x==0 else "yes")

extract_df.shape
```

```python
# Extract rows for only those who attended a bootcamp
attend_df = extract_df.loc[extract_df["AttendedBootcamp"] == "yes", :]
rec_df = extract_df.loc[extract_df["BootcampRecommend"] == "yes", :]
attend_df.head()
```

```python
# Create DataFrame of the different boot camps that had a significant number of attendees
attend_df.groupby("BootcampName", group_keys = True)
rec_df.groupby("BootcampName", group_keys = True)
```

```python
# Count how many attendees of each bootcamp would recommend the bootcamp
bc_attend = attend_df["BootcampName"].value_counts()
bc_attend_df = pd.DataFrame(bc_attend)
bc_attend_df.reset_index(inplace = True)
bc_attend_df.columns = ["Bootcamp Name", "Total Attendance"]

bc_rec = rec_df["BootcampName"].value_counts()
bc_rec_df = pd.DataFrame(bc_rec)
bc_rec_df.reset_index(inplace = True)
bc_rec_df.columns = ["Bootcamp Name", "Total Recommendations"]
```

```python
# Merge the two created data frames on the name of tbe bootcamp
merged_table = pd.merge(bc_attend_df, bc_rec_df, on="Bootcamp Name", how='inner')
merged_table
merged_table["Total Recommendations"].value_counts()
```

```python
# Calculate percentage of each bootcamp's students who are recommenders
merged_table["Pct That Recommend"] = 100 * merged_table["Total Recommendations"] / merged_table["Total Attendance"]

merged_table = round(merged_table.sort_values(by = ["Pct That Recommend"], ascending = False), 2)
merged_table["Pct That Recommend"] = merged_table["Pct That Recommend"].map("{0:,.2f}%".format)
```
