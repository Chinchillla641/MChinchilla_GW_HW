```python
# set dependencies
import pandas as pd
import numpy as np
import matplotlib as plt
import requests
import re as re
```

```python
gdp_path = "D:/Git/MChinchilla_GW_HW/07_Project_01/Data/currency-exchange-master/data_cleaning/gdp.csv"
gdp = pd.read_csv(gdp_path)
gdp
```

```python
chn = gdp.iloc[0, 2:13]
chn = pd.DataFrame(chn)
chn.reset_index(inplace = True)
chn.columns = ("date", "gpd")
chn["year"] = chn["date"].str.extract("([0-9][0-9][0-9][0-9])", expand = True)
chn = chn.iloc[:, 1:3]

jpn = gdp.iloc[1, 2:13]
jpn = pd.DataFrame(jpn)
jpn.reset_index(inplace = True)
jpn.columns = ("date", "gpd")
jpn["year"] = jpn["date"].str.extract("([0-9][0-9][0-9][0-9])", expand = True)
jpn = jpn.iloc[:, 1:3]

usa = gdp.iloc[2, 2:13]
usa = pd.DataFrame(usa)
usa.reset_index(inplace = True)
usa.columns = ("date", "gpd")
usa["year"] = usa["date"].str.extract("([0-9][0-9][0-9][0-9])", expand = True)
usa = usa.iloc[:, 1:3]

gbr = gdp.iloc[3, 2:13]
gbr = pd.DataFrame(gbr)
gbr.reset_index(inplace = True)
gbr.columns = ("date", "gpd")
gbr["year"] = gbr["date"].str.extract("([0-9][0-9][0-9][0-9])", expand = True)
gbr = gbr.iloc[:, 1:3]

usa.columns = ["GDP", "Year"]
usa['GDP'] = usa['GDP'].astype(np.int64)
usa['Year'] = usa['Year'].astype(np.int64)

usa.dtypes
```
